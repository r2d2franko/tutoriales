```python
#Librerias
import pandas as pd
import glob
import os
import datetime

#Funciones de usuario
import source.aux_functions as fn
```

```python
# 1.- Crear Ruta dinamica para archivos CSV
    # La ruta se construye con el año actual y la carpeta de datos de ventas
    # Es mejor usar variables dinamicas para ejecuciones futuras sin modificar el codigo   
    # Puedes añador TRY y EXCEPT para manejar errores en la ruta o archivos no encontrados

anio_actual = datetime.datetime.now().strftime("%Y")
ruta_archivos = f"./datos/datos_ventas/{anio_actual}/"

print(ruta_archivos)

# 2.- Obtener lista de archivos CSV en la ruta especificada,filtrando por nombre comodin y extensión .csv
    # Esta funcion glob es la mas optimizada para buscar archivos en una carpeta con varios archivos csv
    # Es preferible usar archivos de extension parquet o algun otro formato binario para mejorar el rendimiento

archivos_csv = glob.glob(os.path.join(ruta_archivos, "ventas_*.csv"))
df_total = pd.concat(map(fn.leer_archivo, archivos_csv), ignore_index=True)    
df_total
```

```python
# Crear la carpeta si no existe
output_folder = f"./datos/datos_ventas_agrupados/{anio_actual}/"
os.makedirs(output_folder, exist_ok=True)

df_total.to_csv(f"./datos/datos_ventas_agrupados/{anio_actual}/ventas_total.csv", index=False)
```
